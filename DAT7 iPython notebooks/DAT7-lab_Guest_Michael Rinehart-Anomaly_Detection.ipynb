```python
%matplotlib inline
import numpy
import scipy
import scipy.stats
import sklearn.covariance
import sklearn.cluster
import sklearn.preprocessing
from matplotlib import pyplot as plt
```

# Example #1: Detecting Single-Merchant Fraudulent Purchases

## Normal Data Generation

```python
n_customers = [50, 300, 150]  # number of customers who arrive during morning, mid-day, and evening hours
p_purchase = [0.2, 0.7, 0.5]  # probability that a customer makes a purchase at the different time frames of the day
n_days = 120  # number of days in history

# Each index of the each array is the day, and each element are the number of purchases made that day during that part of the day
early_history = numpy.random.binomial(n_customers[0], p_purchase[0], n_days)
midday_history = numpy.random.binomial(n_customers[1], p_purchase[1], n_days)
evening_history = numpy.random.binomial(n_customers[2], p_purchase[2], n_days)

# All-day purchasing history (losing information about when purchases occurred)
allday_history = early_history + midday_history + evening_history

plt.hist(allday_history, 50)
plt.legend(['All Day'])
plt.show()

plt.hist(early_history)
plt.hist(midday_history)
plt.hist(evening_history)
plt.legend(['Morning', 'Mid-Day', 'Evening'])
plt.show()
```

## Visually Detection of Anomalous Behavior

```python
n_fraud_purchases = 15

# Draw a binaomial RV for the number of normal purchases made on "Fraud Day" but add fraudulent purchases only to the morning slot
early_fraud_day = numpy.random.binomial(n_customers[0], p_purchase[0], 1) + n_fraud_purchases
midday_fraud_day = numpy.random.binomial(n_customers[1], p_purchase[1], 1)
evening_fraud_day = numpy.random.binomial(n_customers[2], p_purchase[2], 1)

# Total number of purchases made on "Fraud Day"
allday_fraud_day = early_fraud_day + midday_fraud_day + evening_fraud_day

plt.plot([allday_fraud_day, allday_fraud_day], [0, 10], 'r-')
plt.hist(allday_history, 50)
plt.legend(['Fraud Day', 'History'])
plt.show()

plt.plot([early_fraud_day, early_fraud_day], [0, 30])
plt.plot([midday_fraud_day, midday_fraud_day], [0, 30])
plt.plot([evening_fraud_day, evening_fraud_day], [0, 30])
plt.hist(early_history)
plt.hist(midday_history)
plt.hist(evening_history)
plt.legend(['Morning (Fraud)', 'Mid-Day (Fraud)', 'Evening (Fraud)', 'Morning (Norm)', 'Mid-Day (Norm)', 'Evening (Norm)'])
plt.show()
```

## Programmatic Finding of Fraud

```python
def is_fraudulent(history, x, p_thresh):
    # Convert x from a list to a scalar
    x = x[0]
    
    # Approximate our data with a Gaussian RV
    mean = numpy.mean(history)
    std = numpy.std(history)
    
    # 1-CDF = probability of # purchases exceeding 'x' on a normal day
    p_value = 1 - scipy.stats.norm.cdf(x, mean, std)  
    
    # We'll show our result as the title of the plot but plotting isn't necessary
    domain = numpy.linspace(min(history*0.9), max(history*1.1), 100)
    gaussian_curve = scipy.stats.norm.pdf(domain,mean, std)
    plt.plot(domain, gaussian_curve, [x, x], [0, 0.2])
    
    min_x = min(min(domain), x) * 0.9
    max_x = max(max(domain), x) * 1.1
    min_y = 0
    max_y = 0.2
    plt.axis((min_x, max_x, min_y, max_y))
    
    if p_value < p_thresh:
        plt.title('Detected as Anomalous: p-value = %0.2f' % p_value)
    else:
        plt.title('Detected as Not Anomalous: p-value = %0.2f ' % p_value)
    
    plt.show()
    
is_fraudulent(allday_history, allday_fraud_day, 0.001)
is_fraudulent(early_history, early_fraud_day, 0.001)
is_fraudulent(midday_history, midday_fraud_day, 0.001)
is_fraudulent(evening_history, evening_fraud_day, 0.001)
```

# Example #2: Detecting Fraudulent Merchants as Outliers: Normally-Distributed Data

## Normal Data Generation

```python
n_merchants = 500

# First, we will draw fake data for the number of customers each merchant saw that morning
avg_n_customers = n_customers[0]  # In our exmple, all merchants have the same average number of customers in the morning
std_n_customers = 0.3 * avg_n_customers  # Apply a small standard deviation
n_customers_per_merchant = numpy.zeros(n_merchants)
for merchant in range(n_merchants):
    n = numpy.random.normal(avg_n_customers, std_n_customers)
    n_customers_per_merchant[merchant] = max(int(n), 0)  # no "negative" customer counts
    
# Next, we'll draw fake data for how much each customer spent and accumulate to get the total morning sales. 
# We'll again assume an average with a standard deviation.
avg_expenditure = 7.00
std_expenditure = 0.4 * avg_expenditure
sales_per_merchant = numpy.zeros(n_merchants)
for merchant in range(n_merchants):
    for sale in range(int(n_customers_per_merchant[merchant])):
        sale_amount = numpy.random.normal(avg_expenditure, std_expenditure)
        sales_per_merchant[merchant] += max(sale_amount, 0.0)  # all sales are final
        
plt.plot(n_customers_per_merchant, sales_per_merchant, 'b.')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')
plt.show()
```

## Robust Level-Set (Co-Variance) Estimation on Normal Data

```python
def add_level_sets(robust_cov, emp_cov):
    # Level-set plotting is well done here:
    #     http://scikit-learn.org/stable/auto_examples/covariance/plot_mahalanobis_distances.html#example-covariance-plot-mahalanobis-distances-py
    xx, yy = numpy.meshgrid(numpy.linspace(plt.xlim()[0], plt.xlim()[1], 100),
                            numpy.linspace(plt.ylim()[0], plt.ylim()[1], 100))
    zz = numpy.c_[xx.ravel(), yy.ravel()]
    mahal_emp_cov = emp_cov.mahalanobis(zz)
    mahal_emp_cov = mahal_emp_cov.reshape(xx.shape)
    emp_cov_contour = plt.contour(xx, yy, numpy.sqrt(mahal_emp_cov),
                                      cmap=plt.cm.PuBu_r,
                                      linestyles='dashed')

    mahal_robust_cov = robust_cov.mahalanobis(zz)
    mahal_robust_cov = mahal_robust_cov.reshape(xx.shape)
    robust_contour = plt.contour(xx, yy, numpy.sqrt(mahal_robust_cov),
                                     cmap=plt.cm.YlOrBr_r, linestyles='dotted')

    plt.show()
```

```python
normal_data = numpy.vstack((n_customers_per_merchant, sales_per_merchant)).transpose()  # Make a single, 2-D dataset
robust_cov = sklearn.covariance.EllipticEnvelope().fit(normal_data)  # Robust co-variance
emp_cov = sklearn.covariance.EmpiricalCovariance().fit(normal_data)  # Empirical co-variance (sensitive to outliers)

plt.plot(normal_data[:,0], normal_data[:,1], 'b.')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')

add_level_sets(robust_cov, emp_cov)
```

## Visual Inspection of Outliers

```python
# Now make it so the first merchant committed fraud
fraud_data = normal_data.copy()  # do not want to overwrite previous data
n_extra_customers = avg_n_customers  
fraud_data[0, 0] += n_extra_customers
fraud_data[0, 1] += n_extra_customers * (avg_expenditure * 2)  # Fake sales are, on average, 50% higher
fraud_data[1, 0] += n_extra_customers * 2
fraud_data[1, 1] += n_extra_customers * (avg_expenditure)  # Fake sales are, on average, 50% higher

# Construct co-variances (level-set geometries) for this data using both robust and standard (not robust) estimation
robust_cov = sklearn.covariance.EllipticEnvelope().fit(fraud_data)  # Robust co-variance
emp_cov = sklearn.covariance.EmpiricalCovariance().fit(fraud_data)  # Empirical co-variance (sensitive to outliers)

plt.plot(fraud_data[2:,0], fraud_data[2:,1], 'b.')
plt.plot(fraud_data[0:2,0], fraud_data[0:2,1], 'ro')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')

add_level_sets(robust_cov, emp_cov)
```

## Programmatic Detection of Outliers

```python
def outlier_plot(prediction_per_merchant):
    for merchant in range(n_merchants):
        x = fraud_data[merchant, :]
        if prediction_per_merchant[merchant] > 0:
            plt.plot(x[0], x[1], 'b.')
        else:
            plt.plot(x[0], x[1], 'ro')
        
    plt.title('# Customers vs Sales during 7am-10am on 8/9/2014')
    plt.xlabel('# customers')
    plt.ylabel('Sales ($)')
    plt.show()
```

```python
prediction_per_merchant = robust_cov.predict(fraud_data)
outlier_plot(prediction_per_merchant)
```

```python
contamination_fraction = 0.003  # Experimentally-determined. We'll try different values
tuned_robust_cov = sklearn.covariance.EllipticEnvelope(contamination=contamination_fraction).fit(fraud_data)
prediction_per_merchant = tuned_robust_cov.predict(fraud_data)

outlier_plot(prediction_per_merchant)
```

# Example #3: Detecting Fraudulent Merchants as Outliers: Nonnormally-Distributed Data

## Normal Data Generation

```python
n_merchants = 1000

# First, we will draw fake data for the number of customers each merchant saw that morning
avg_n_customers = numpy.array([n_customers[0]*0.5, n_customers[0]*5])  # several averages
std_n_customers = 0.1 * avg_n_customers  # Standard deviations will scale with the averages
n_customers_per_merchant = numpy.zeros(n_merchants)
for merchant in range(n_merchants):
    group = numpy.random.randint(0, 2)
    n = numpy.random.normal(avg_n_customers[group], std_n_customers[group])
    n_customers_per_merchant[merchant] = max(int(n), 0)  # no "negative" customer counts
    
# Next, we'll draw fake data for how much each customer spent and accumulate to get the total morning sales. 
# We'll again assume an average with a standard deviation.
avg_expenditure = numpy.array([7.00, 5.00, 2.50])
std_expenditure = 0.4 * avg_expenditure
sales_per_merchant = numpy.zeros(n_merchants)
for merchant in range(n_merchants):
    group = numpy.random.randint(0, 3)
    for sale in range(int(n_customers_per_merchant[merchant])):
        sale_amount = numpy.random.normal(avg_expenditure[group], std_expenditure[group])
        sales_per_merchant[merchant] += max(sale_amount, 0.0)  # all sales are final
        
plt.plot(n_customers_per_merchant, sales_per_merchant, 'b.')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')
plt.show()
```

## DBSCAN on Normal Data

```python
def plot_dbscan(db, data):
    # See http://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html
    core_samples_mask = numpy.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    colors = plt.cm.Spectral(numpy.linspace(0, 1, len(unique_labels)))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = 'k'

        class_member_mask = (labels == k)

        xy = data[class_member_mask & core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=14)

        xy = data[class_member_mask & ~core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=6)

    plt.title('# Clusters: %d; # Outliers = %d' % (n_clusters_, sum(labels == -1)))
    plt.show()
```

```python
normal_data = numpy.vstack((n_customers_per_merchant, sales_per_merchant)).transpose()  # Make a single, 2-D dataset
normalized_data = sklearn.preprocessing.StandardScaler().fit_transform(normal_data)

plt.plot(normalized_data[:,0], normalized_data[:,1], 'b.')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014 (Normalized)')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')
plt.show()

db = sklearn.cluster.DBSCAN(eps=0.1, min_samples=10).fit(normalized_data)
plot_dbscan(db, normalized_data)

db = sklearn.cluster.DBSCAN(eps=0.1, min_samples=2).fit(normalized_data)
plot_dbscan(db, normalized_data)

db = sklearn.cluster.DBSCAN(eps=0.5, min_samples=10).fit(normalized_data)
plot_dbscan(db, normalized_data)
```

## Visual and Programmatic Detection of Outliers

```python
fraud_data = normal_data.copy()  # do not want to overwrite previous data
n_extra_customers = 30  # Effectively runs 25 credit cards that morning
fraud_data[0, 0] *= 2
fraud_data[0, 1] *= 2
fraud_data[1, 0] *= 2
fraud_data[1, 1] *= 2

normalized_fraud_data = sklearn.preprocessing.StandardScaler().fit_transform(fraud_data)

plt.plot(normalized_fraud_data[:,0], normalized_fraud_data[:,1], 'b.')
plt.title('# Customers vs Sales during 7am-10am on 8/9/2014 (Normalized)')
plt.xlabel('# customers')
plt.ylabel('Sales ($)')
plt.show()

db = sklearn.cluster.DBSCAN(eps=0.1, min_samples=10).fit(normalized_fraud_data)
plot_dbscan(db, normalized_fraud_data)

db = sklearn.cluster.DBSCAN(eps=0.1, min_samples=2).fit(normalized_fraud_data)
plot_dbscan(db, normalized_fraud_data)

db = sklearn.cluster.DBSCAN(eps=0.5, min_samples=10).fit(normalized_fraud_data)
plot_dbscan(db, normalized_fraud_data)
```
