```python
#usual imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import cross_validation
%matplotlib inline

# Statsmodel APIs
import statsmodels.api as sm
```

```python
#set up a data prep loop that inverts to the "inversPower" value

def dataPrep(inversePower, data, test_size, random_state):
    #set up data
   
    data = [float(i) for i in data]

    # create inverted x data 
    x = range(1,len(data)+1)
    x = [pow(i, inversePower) for i in x]

    # add intercepts column 
    X = np.c_[x, np.ones(len(data))]
    
    return cross_validation.train_test_split(X, data, test_size = test_size, random_state=random_state)
```

```python
data = [37728, 12571, 3476, 1740, 868, 532, 391, 335, 255, 207, 175, 161, 137, 80, 103, 114, 92, 87, 105, 67, 79, 58, 57, 47, 58, 57, 36, 39, 39, 33, 34, 37, 34, 29, 45, 42, 34, 27, 32, 26, 20, 23, 21, 24, 19, 18, 17, 12, 21, 9, 14, 13, 18, 15, 13, 14, 21, 15, 7, 11]


# training model complexity
R_squared_adj_results = list()
AIC_results = list()
BIC_results = list()

#out of sample accuracy

for i in range(5):
    X_train, X_test, y_train, y_test = dataPrep(inversePower = -i, data = data, test_size = 0.20, random_state=1 )
  
    results = sm.OLS(y_train, X_train).fit()
    print "inverse power of " + str(i)
    print "    Adjusted R-squared is " + str(results.rsquared_adj)
    R_squared_adj_results.append(results.rsquared_adj)
    
    print "    AIC is "+str(results.aic)
    AIC_results.append(results.aic)
    
    print "    BIC is "+str(results.bic)
    BIC_results.append(results.bic)
```

```python
AIC_results_Array = np.array(AIC_results)
BIC_results_Array = np.array(BIC_results)

# print max(AIC_results)
# normalize by max value and invert AIC and BIC results
AIC_results_neg_norm = AIC_results_Array/max(AIC_results)
BIC_results_neg_norm = BIC_results_Array/max(BIC_results)

# max point of the BLUE line is optimal
plt.plot(R_squared_adj_results, 'b-')
#print R_squared_adj_results

# max point of the dashed RED line is optimal
plt.plot(AIC_results_neg_norm, 'r--')
#print AIC_results_neg_norm

# max point of the dotted GREEN line is optimal
plt.plot(BIC_results_neg_norm, 'g-.')
#print BIC_results_neg_norm
```

```python
def findOptimum(R_squared_adj_results, AIC_results, BIC_results):
    if np.argmax(R_squared_adj_results) == np.argmin(AIC_results):
        if np.argmax(R_squared_adj_results) == np.argmin(BIC_results):
            print "optimal inverse power is -" + str(np.argmax(R_squared_adj_results))
            return -np.argmax(R_squared_adj_results)
        else:
            print "WARNNING: BIC optimum does not match Adjusted R-squared optimum"
    else:
            print "WARNNING: AIC optimum does not match Adjusted R-squared optimum \n"+ "Consider using Adjusted R-squared optimum power of -" + str(np.argmax(R_squared_adj_results))
            
optimal_inverse_power = findOptimum(R_squared_adj_results, AIC_results, BIC_results)
```

```python
# Now we create the optimal model using optimal_inverse_power for testing 
# to confirm that optimal_inverse_power is the best, we calculate for a range of interger dviations -1 to 4

for i in range(-2,5):
    
    X_train, X_test, y_train, y_test = dataPrep(inversePower = optimal_inverse_power+i, data = data, test_size = 0.30, random_state=1)
    optimal_results = sm.OLS(y_train, X_train).fit()

    # predict values at test slots using the model
    y_hat = optimal_results.predict(X_test)
    plt.plot(y_hat)
    #print y_hat
    plt.plot(y_test)
    #print y_test

    res = y_hat-y_test
    logMSE = np.log(sum(res**2)/len(res))
    #print res
    print "for our optimal inverse power of \n    " + str(optimal_inverse_power) + " + " + str(-i) + "\n    the log mean squared error is \n    " + str(logMSE) + "\n"
    
```

```python
# lotting for the optimal model alone using optimal_inverse_power for testing 
X_train, X_test, y_train, y_test = dataPrep(inversePower = optimal_inverse_power, data = data, test_size = 0.30, random_state=1)
optimal_results = sm.OLS(y_train, X_train).fit()

# predict values at test slots using the model
y_hat = optimal_results.predict(X_test)
plt.plot(y_hat)
#print y_hat
plt.plot(y_test)
#print y_test

res = y_hat-y_test
logMSE = np.log(sum(res**2)/len(res))
print "for our optimal inverse power of \n    " + str(optimal_inverse_power) + "\n    the log mean squared error is \n    " + str(logMSE) + "\n"


```
